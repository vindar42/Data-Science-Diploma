```python
import requests
from bs4 import BeautifulSoup

url = "https://www.exchangerates.org.uk/Dollars-to-Egyptian-Pounds-currency-conversion-page.html"

response = requests.get(url)
print(response.text)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
soup
```

```python
usd = soup.findChildren("div", {"class": "p_conv30"})[1].get_text().split()
usd
```

```python
def exchange_USD():
    integer = float(usd[3])
    user = input("Please Enter USD: ")
    num = float(user) * integer
    return f"you have {num} EGP"

exchange_USD()
```
