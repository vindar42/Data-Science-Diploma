```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(rc={'figure.figsize':(10, 10)}, font_scale=1.2)
```

```python
df = pd.read_csv('../dastasets/ufo.csv')
df.head()
```

```python
df['Time'] = pd.to_datetime(df['Time'], format='%m/%d/%Y %H:%M', errors="coerce")
df.head()
```

```python
def in_december(date):
    if date.month == 12:
        return 1
    else:
        return 0
    
df['IN_December'] = df['Time'].apply(in_december)
```

```python
from datetime import datetime
df['Time_Diff_From_WW1'] = (pd.to_datetime(df['Time']) - pd.to_datetime('1914-7-28')) / np.timedelta64(1, 'Y')
df.drop('diff', axis= 1)
```
