```python
df = pd.read_csv('../dastasets/Restaurant_Reviews.csv', sep='\t')
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(stop_words='english')
```

```python
vectorizer.fit(df['Review'])
vectorizer.vocabulary_
```

```python
vectorizer.get_feature_names()
```

```python
Review_vecctor = vectorizer.fit_transform(df['Review'])
Review_vecctor.toarray()
```

```python
df = pd.DataFrame(Review_vecctor.toarray(), columns=vectorizer.get_feature_names())
df
```

```python
df = pd.read_csv('../dastasets/Restaurant_Reviews.csv', sep='\t')
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english')
```

```python
vectorizer.fit(df['Review'])

vectorizer.vocabulary_
```

```python
vector = vectorizer.transform(df['Review'])

vector.toarray()
```

```python
import pandas as pd

tfidf_df = pd.DataFrame(vector.toarray(), columns=vectorizer.get_feature_names())
tfidf_df
```
