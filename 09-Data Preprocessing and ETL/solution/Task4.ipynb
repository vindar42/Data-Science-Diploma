```python
import user_agents

def extract_UserAgent(ua):
    ua = user_agents.parse(ua)
    return pd.Series((ua.os.family, ua.browser.family, ua.device.family))
    
df[['OS', 'Browser', 'Device']] = df['UserAgent'].apply(extract_UserAgent)
df.head(50)
```

```python
df = pd.read_csv('../dastasets/visitor-interests.csv')
df.head()
```

```python
from ip2geotools.databases.noncommercial import DbIpCity as ip2geo

def extract_ip(x):
    try:
        response = ip2geo.get(x, api_key='free')
        return pd.Series((response.latitude, response.longitude))
    except:
        return pd.Series((np.nan, np.nan))

df[['lat', 'long']] = df['IP'].apply(extract_ip)
df.head(5)
```
